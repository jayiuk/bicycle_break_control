```python
use12 = pd.read_csv('C:/bicycle_data/2312.csv', encoding = 'cp949')
use11 = pd.read_csv('C:/bicycle_data/2311.csv', encoding = 'cp949')
use10 = pd.read_csv('C:/bicycle_data/2310.csv', encoding = 'cp949')
use9 = pd.read_csv('C:/bicycle_data/2309.csv', encoding = 'cp949')
use8 = pd.read_csv('C:/bicycle_data/2308.csv', encoding = 'cp949')
use7 = pd.read_csv('C:/bicycle_data/2307.csv', encoding = 'cp949')
use6 = pd.read_csv('C:/bicycle_data/2306.csv', encoding = 'cp949')
use5 = pd.read_csv('C:/bicycle_data/2305.csv', encoding = 'cp949')
use4 = pd.read_csv('C:/bicycle_data/2304.csv', encoding = 'cp949')
use3 = pd.read_csv('C:/bicycle_data/2303.csv', encoding = 'cp949')
use2 = pd.read_csv('C:/bicycle_data/2302.csv', encoding = 'cp949')
use1 = pd.read_csv('C:/bicycle_data/2301.csv', encoding = 'cp949')
use12
```
