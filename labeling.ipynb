```python
bike_break = bike_break.rename(columns = {'등록일시' : 'date'})
bike_break.date = bike_break.date.str[:10]
bike_break['date'] = pd.to_datetime(bike_break['date'])
bike_break
```

```python
bike_break_sum = bike_break.groupby('date').count()
bike_break_sum
```

```python
bike_break_sum = bike_break_sum.reset_index()
bike_break_sum = bike_break_sum.drop(bike_break_sum.columns[1], axis = 1)
bike_break_sum
```

```python
bike_use_break = pd.merge(bike_use_sum, bike_break_sum, on = 'date')
bike_use_break
```

```python
bike_use_break.corr()
```

```python
bike_break_c = bike_use_break.copy()
bike_break_c = bike_break_c.drop(['대여소번호', 'date'], axis = 1)
bike_break_c
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
km = KMeans(n_clusters = 4, max_iter = 20, random_state = 42)
data = bike_break_c
```

```python
km.fit(data)
labels = km.labels_
```

```python
km_break = bike_break_c.copy()
```

```python
km_break['cluster'] = labels
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(bike_break_c)
pca_transformed
```

```python
len(pca_transformed)
```

```python
km_break['pca_x'] = pca_transformed[:, 0]
km_break['pca_y'] = pca_transformed[:, 1]
km_break
```

```python
w0 = km_break[km_break['cluster'] == 0].index
w1 = km_break[km_break['cluster'] == 1].index
w2 = km_break[km_break['cluster'] == 2].index
w3 = km_break[km_break['cluster'] == 3].index
```

```python
plt.scatter(x = km_break.loc[w0, 'pca_x'], y = km_break.loc[w0, 'pca_y'], color = 'black')
plt.scatter(x = km_break.loc[w1, 'pca_x'], y = km_break.loc[w1, 'pca_y'], color = 'red')
plt.scatter(x = km_break.loc[w2, 'pca_x'], y = km_break.loc[w2, 'pca_y'], color = 'yellow')
plt.scatter(x = km_break.loc[w3, 'pca_x'], y = km_break.loc[w3, 'pca_y'], color = 'purple')
plt.show()
```
